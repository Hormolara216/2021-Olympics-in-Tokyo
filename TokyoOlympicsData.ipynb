```
Athlete_data=spark.read.format("csv").option("header", "true").load("/mnt/2021olympics/raw_data/Athletes.csv")
Coaches_data=spark.read.format("csv").option("header", "true").load("/mnt/2021olympics/raw_data/Coaches.csv")
EntriesGender_data=spark.read.format("csv").option("header", "true").load("/mnt/2021olympics/raw_data/EntriesGender.csv")
Medals_data=spark.read.format("csv").option("header", "true").option("inferSchema","true").load("/mnt/2021olympics/raw_data/Medals.csv")
Teams_data=spark.read.format("csv").option("header", "true").option("inferSchema","true").load("/mnt/2021olympics/raw_data/Teams.csv")
```

```
#change the datatype of Female, Male and Total to Int
EntriesGender_data=EntriesGender_data.withColumn("Female", col("Female").cast(IntegerType())) \
    .withColumn("Male",col("Male").cast(IntegerType())) \
    .withColumn("Total",col("Total").cast(IntegerType()))
```

```
#highest country with the number of gold medals
Medals_data.orderBy("Gold",ascending=False).select("Team/NOC","Gold").show()
```

```
#the average number of entries by gender for each discipline
EntriesGender_data.withColumn("Avg_Female",EntriesGender_data["Female"]/EntriesGender_data["Total"])\
    .withColumn("Avg_Male",EntriesGender_data["Male"]/EntriesGender_data["Total"])\
.show()
```

```
#load transformed data to Data lake
Athlete_data.repartition(1).write.mode("overwrite").option("header","true").csv("mnt/2021olympics/transformed_data/Athlete")
Coaches_data.repartition(1).write.mode("overwrite").option("header","true").csv("mnt/2021olympics/transformed_data/Coaches")
EntriesGender_data.repartition(1).write.mode("overwrite").option("header","true").csv("mnt/2021olympics/transformed_data/EntriesGender")
Medals_data.repartition(1).write.mode("overwrite").option("header","true").csv("mnt/2021olympics/transformed_data/Medals")
Teams_data.repartition(1).write.mode("overwrite").option("header","true").csv("mnt/2021olympics/transformed_data/Teams")
```
